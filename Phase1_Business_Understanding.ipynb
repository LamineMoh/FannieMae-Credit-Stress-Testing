### CRISP-DM Phase 1: Define Business Objectives and Success Criteria

## 1.1 Business Objective

**Goal**: Predict mortgage loan defaults during Q1 2008 (peak of financial crisis)

### Why This Matters
- **Risk Management**: Identify high-risk loans for proactive management
- **Stress Testing**: Evaluate portfolio risk under crisis conditions
- **Regulatory Compliance**: Meet stress testing requirements
- **Capital Planning**: Estimate potential losses

## 1.2 Target Variable Definition

| Delinquency Status | Description | Classification |
|-------------------|-------------|----------------|
| 0 | Current (not delinquent) | **No Default** |
| 1-2 | 30-60 days delinquent | No Default |
| 3+ | 90+ days delinquent | **DEFAULT** |
| RA | REO Acquisition | **DEFAULT** |
| F, S, T, N | Foreclosure/Short Sale | **DEFAULT** |

## 1.3 Success Criteria

| Metric | Target | Importance |
|--------|--------|------------|
| **AUC-ROC** | > 0.70 | Primary - Model discriminative ability |
| **Recall** | > 0.60 | High - Minimize missed defaults |
| **Precision** | > 0.50 | Medium - Reduce false alarms |
| **F1-Score** | > 0.55 | Balance precision/recall |

## 1.4 Key Stakeholders

- **Risk Management Team**: Primary users of default predictions
- **Credit Portfolio Managers**: Loan-level decision making
- **Regulatory Compliance**: Stress test documentation
- **Senior Leadership**: Strategic risk assessment
